```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('F:\\datasets\\zoo_class\\zoo.csv')
df = file.copy()
df.sample(10)
```

```python
label = df['class_type']
df = df.drop(['class_type','animal_name'],axis=1)
df.info()
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(df,label,test_size=0.2,random_state=42)
print('spliting done !')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_xtr = pd.DataFrame(scaler.fit_transform(xtr),columns=df.columns)
scaled_xts = pd.DataFrame(scaler.fit_transform(xts),columns=df.columns)
```

```python
scaled_xtr.info()
```

```python
scaled_xts.info()
```

```python
from sklearn.svm import LinearSVC
alg = LinearSVC()
alg.fit(scaled_xtr,ytr)
print(alg.score(scaled_xtr,ytr)*100)
print(alg.score(scaled_xts,yts)*100)
```

```python
from sklearn.metrics import confusion_matrix
pred = alg.predict(scaled_xts)
confusion_matrix(yts,pred)
```

```python
sacaled_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
alg.fit(sacaled_df,label)
print(alg.score(sacaled_df,label)*100)
```
